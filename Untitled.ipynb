```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("reduced_pollution_dataset.csv")
df.head()
```

### As we can see from the above table that we have a total of 9 independent factors which ccan be used to predict the 1 dependent factor ,i.e, Air Quality.
### We will first try to reduce the no.of dependent factors and check correlation of each with our dependent factor.

```python
df['Air Quality'].value_counts()
```

```python
d1 = {'Good':0,'Moderate':1,'Poor':2,'Hazardous':3}
df['Air Quality'] = df['Air Quality'].map(d1)
```

```python
df.head()
```

### First we mapped the string values of column 'Air Quality' to integer values so that we can draw up a correlation matrix and also dealing with integer values is much easier.

```python
corr = df[['Temperature','Humidity','PM2.5','PM10','NO2','SO2','Proximity_to_Industrial_Areas','Population_Density','Air Quality']].corr()
sns.heatmap(corr, annot=True, cmap='RdBu', fmt=".2f")
plt.title('CORRELATION MATRIX')
plt.show()
```

### As we can see, we cannot disregard any of our independent factors as all of them have a pretty high correlation with our independent factor that we are trying to predict.

```python
count = df['Air Quality'].value_counts()
plt.pie(count,labels=('Good','Moderate','Poor','Hazardous'),explode=(0,0,0,0.2))
plt.title('AIR QUALITY INDEX DIVISION')
plt.show()
```

### From the above graph, we can see that our dataset contains a good division of values across all types of air quality.

```python
df.head()
```

```python
X = df[['Temperature','Humidity','PM2.5','PM10','NO2','SO2','Proximity_to_Industrial_Areas','Population_Density']]
Y = df['Air Quality']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 
```

```python
X_train
```

```python
Y_train
```

### Since it is a classification model, I will  try to train this model using decision trees.

```python
from sklearn import tree

model = tree.DecisionTreeClassifier(max_depth=6, random_state=42)
model = model.fit(X_train,Y_train)
```

```python
features = list(X_train.columns)
features
```

```python
from IPython.display import Image  
from io import StringIO  
import pydotplus

dot_data = StringIO()  
tree.export_graphviz(model, out_file=dot_data,  
                         feature_names=features)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

```python
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score

r2_test = r2_score(Y_test, pred)
r2_test
```

```python
pred1 = model.predict(X_train)
pred1
```

```python
r2_train = r2_score(Y_train,pred1)
r2_train
```

### So, we have successfully created the model and it has a r2 score of 0.91 for testing dataset and a 0.95 score for training dataset.

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, pred)

print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_train, pred1)

print(f"Accuracy: {accuracy:.2f}")
```

### Therefore we have a good model considering the high accuracy scores for both training and testing predictions
